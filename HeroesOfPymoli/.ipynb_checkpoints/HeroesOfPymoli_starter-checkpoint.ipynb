```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
totalplayers=purchase_data.SN.nunique() #this returns number of unique elements in the desired column
```

## Player Count

* Display the total number of players


```python
#creates a Data Frame with one row
TotalPlayersdf = pd.DataFrame(index=[0])
#creates the column and sets totalplayers as column data
TotalPlayersdf['Total Players'] = totalplayers
TotalPlayersdf
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#returns number of unique Items
numofitems=purchase_data['Item Name'].nunique()
#average price of items
avgprice=purchase_data['Price'].mean()
numofpurchases=purchase_data['Purchase ID'].nunique()
#sums the order price to get total revenue
TotalRevenue=purchase_data['Price'].sum()
```

```python
#makes a DF of one row
purchaseanalysisdf=pd.DataFrame(index=[0])
#inputs the values colums and rows
purchaseanalysisdf[['Number of Unique Items','Average Price', 'Number of Purchases','Total Revenue']]=[numofitems, f'${avgprice:,.2f}',numofpurchases,f'${TotalRevenue:,.2f}']
purchaseanalysisdf
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
purchase_data.SN.nunique()
# creates a DF of no repeating screen names(SN) 
noduplicates=purchase_data.drop_duplicates(subset='SN').copy()
#group by gender
genderdemo=noduplicates.groupby('Gender').count()[['SN']]
genderdemo.columns=[['Total Count']]
totalSN=genderdemo['Total Count'].sum()
genderdemo['Percentage of Players']=(genderdemo['Total Count']/totalSN)*100
#genderdemo['Percentage of Players']=genderdemo['Percentage of Players'].map('{:.2f}%'.format)
genderdemo
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchasegender=purchase_data.groupby('Gender').mean()
#converts to a list so i can place in new DataFrame later
averagepurchaseprice=purchasegender['Price']
averagepurchaseprice=averagepurchaseprice.tolist()  

purchasecount=purchase_data.groupby('Gender').count()['Purchase ID']
purchasecount=purchasecount.tolist()

totalpurchasevalue=purchase_data.groupby('Gender').sum()['Price']
totalpurchasevalue=totalpurchasevalue.tolist()

data={'Purchase Count':purchasecount, 'Average Purchase Price':averagepurchaseprice,
      'Total Purchase Value': totalpurchasevalue, 'Total Count Per Gender':[81,484,11]} #dict of lists to create new DF
finalpurchgender=pd.DataFrame(data,index=['Female','Male','Other / Non-Disclosed']) #new dataframe to put my data list variable
#final column calculation
finalpurchgender['Avg Total Purchase per Person']=(finalpurchgender['Total Purchase Value']/finalpurchgender['Total Count Per Gender'])
#Format Columns dealing with money
finalpurchgender['Average Purchase Price']=finalpurchgender['Average Purchase Price'].map('${:,.2f}'.format)
finalpurchgender['Total Purchase Value']=finalpurchgender['Total Purchase Value'].map('${:,.2f}'.format)
finalpurchgender['Avg Total Purchase per Person']=finalpurchgender['Avg Total Purchase per Person'].map('${:,.2f}'.format)

finalpurchgender
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
purchase_data.head()
```

```python
#using bins/pd.cut() method to break them into Age groups
bins=[0,9.9,14.9,19.9,24.9,29.9,34.9,39.9,50]
labels=['<10','10-14','15-19','20-24','25-29','30-34','35-39','40+']
#using table with no repeating SN
noduplicates['Age Groups']=pd.cut(noduplicates['Age'], bins, labels=labels)
agedemo=noduplicates.groupby('Age Groups').count()[['Age']]
#calculates the percentage of players
agedemo['Percentage of Players']=(agedemo['Age']/576)*100
agedemo['Percentage of Players']=agedemo['Percentage of Players'].map('{:.2f}%'.format) #formatting 
agedemo
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
bins=[0,9.9,14.9,19.9,24.9,29.9,34.9,39.9,50]
labels=['<10','10-14','15-19','20-24','25-29','30-34','35-39','40+']
purchase_data['Age Groups']=pd.cut(purchase_data['Age'], bins, labels=labels)
purchase_data.groupby('Age Groups').count()['SN']
```

```python
grp=purchase_data.groupby("Age Groups") #using groupby to utilize binning performing
```

```python
#new DF 
df2=pd.DataFrame() 
#fill the columns up
df2['Purchase Count']=grp.SN.count()
df2['Average Purchase Price']=grp.Price.mean()
df2['Total Purchase Value']=grp.Price.sum()
df2['Avg Total Purchase per Person']=(grp.Price.sum()/agedemo.Age)
df2['Avg Total Purchase per Person']=df2['Avg Total Purchase per Person'].map('${:.2f}'.format)

df2
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
pd.options.display.float_format = '${:,.2f}'.format
```

```python
grp = purchase_data.groupby('SN').Price
#creating DF {col:values}
TopSpenders= pd.DataFrame({
    'Purchase Count': grp.count(),
    'Average Purchase Price': grp.mean(),
    'Total Purchase Value': grp.sum()
}).sort_values('Total Purchase Value', ascending=False)
TopSpenders.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
grp2=purchase_data.groupby(['Item ID','Item Name']).Price
```

```python
#creating DF {col:values}
PopularItems=pd.DataFrame({
    'Purchase Count':grp2.count(),
    'Item Price':grp2.mean(),
    'Total Purchase Value': grp2.sum()
}).sort_values('Purchase Count', ascending=False)
PopularItems.head()
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
#sorted by Total Purchase Value and displayed top 5
MostProfitableItems = PopularItems.sort_values('Total Purchase Value', ascending=False)
MostProfitableItems.head()
```
